# **Let's introduce a variable**

A variable is a name which store a value (like a number).

```python
a = 4

print(a)
```

Notice how it didn't print a, but printed 4. If you change the number, it will print out that new number as well (try it out)

We can do math now, but using variables:

```python
a = 4
b = 3

print(a + b)
```

Notice that you can change the numbers, and the code still works. This is why variables can be useful. Later on, we will show that computers can store many numbers or even images in variable names.

# **Strings and Characters**

Other than numbers, computers also work with human language. We call these letters as "strings" or "characters" to differentiate them from numbers. Dealing with letters is a bit different: you can still do operations to language, but its a lot harder.

```python
A = "Hello"
B = "John"

print(A + B)
```

That + symbol put those two words together, but we need to be more careful as we can't do things to the variables filled with strings that we can with those filled with numbers.

```python
A = "Hello"
B = "John"
a = 5
b = 4

print(a + b)
print(A + B)
print(a - b)
print(A - B) # this gives an error
```

You have to be careful in code, the computer needs to recognize what you are writing.
