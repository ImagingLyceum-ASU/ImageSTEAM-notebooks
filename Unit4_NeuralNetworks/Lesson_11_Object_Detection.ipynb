# **Activity: Template matching**


```
#@markdown We use something called template matching to find an object automatically in a picture.

#@markdown Can you spot Waldo in the image. Using the number grid in the above picture, coarsely estimate where Waldo is and enter the values below. This is used a template to search for Waldo by the program.

waldo = crop(image)

templateMatch(image=image, template=waldo)
```

With that the program has found Waldo/Wally using the template you provided! and drawn a green rectangle drawn around him!



```
#@markdown Now let's try to find with another template of Waldo, and try to do the matching. Does it work better than your template?

waldo2 = data.waldo2()[:,:,:3]
templateMatch(image, template=waldo2)
```
